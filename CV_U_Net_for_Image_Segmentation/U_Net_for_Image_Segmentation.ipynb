```python
"""
U-Net for Image Segmentation
Author: [Rhishi Kumar Ayyappan]

- Installs all necessary libraries (tensorflow, tensorflow-datasets).
- Loads the 'oxford_iiit_pet' dataset (version 4.0.0), which downloads automatically (no API key needed).
- Preprocesses the images and segmentation masks.
- Builds a complete U-Net model from scratch using tf.keras.
- Trains the model on the dataset.
- Visualizes the results, showing the original image, the true mask, and the model's predicted mask side-by-side.
"""

# =============================
# SECTION 1: Install Libraries & Imports
# =============================
# This cell installs/updates the required packages
!pip install -q tensorflow tensorflow-datasets seaborn

import os
import warnings
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow.keras import layers, models, losses, optimizers
from IPython.display import display, Image, HTML
# from google.colab.files import FileLink # Import FileLink - THIS LINE CAUSES ERROR
import seaborn as sns # Import seaborn

# Suppress warnings
warnings.filterwarnings("ignore")

# Set a modern, consistent, colorblind-friendly style
sns.set_theme(style="whitegrid", palette="deep", font_scale=1.1)
plt.rcParams.update({
    'axes.titlesize': 16,
    'axes.labelsize': 13,
    'xtick.labelsize': 11,
    'ytick.labelsize': 11,
    'axes.titleweight': 'bold',
    'legend.fontsize': 11,
    'axes.edgecolor': '#AAAAAA',
    'axes.linewidth': 1.2
})


# =============================
# SECTION 2: Data Download & Preprocessing
# =============================

print("Loading 'oxford_iiit_pet:4.0.0' dataset from TensorFlow Datasets...")
# This will automatically download and prepare the dataset
dataset, info = tfds.load('oxford_iiit_pet:4.0.0', with_info=True)

# --- Define Preprocessing Functions ---

IMG_SIZE = 128

def normalize(input_image, input_mask):
    """Normalizes images to [0, 1] and resizes them."""
    input_image = tf.image.resize(input_image, (IMG_SIZE, IMG_SIZE))
    input_mask = tf.image.resize(input_mask, (IMG_SIZE, IMG_SIZE), method='nearest')

    input_image = tf.cast(input_image, tf.float32) / 255.0

    # Segmentation masks are 1, 2, 3. We subtract 1 to make them 0, 1, 2
    # 0=Pet, 1=Border, 2=Background
    input_mask -= 1

    return input_image, input_mask

@tf.function
def load_image_train(datapoint):
    """Loads and augments the training image and mask."""
    input_image = datapoint['image']
    input_mask = datapoint['segmentation_mask']

    # Simple augmentation: random flip
    if tf.random.uniform(()) > 0.5:
        input_image = tf.image.flip_left_right(input_image)
        input_mask = tf.image.flip_left_right(input_mask)

    input_image, input_mask = normalize(input_image, input_mask)
    return input_image, input_mask

def load_image_test(datapoint):
    """Loads and normalizes the test image and mask."""
    input_image = datapoint['image']
    input_mask = datapoint['segmentation_mask']
    input_image, input_mask = normalize(input_image, input_mask)
    return input_image, input_mask

# --- Create Data Pipelines ---

BATCH_SIZE = 64
BUFFER_SIZE = 1000

# Training pipeline
train_dataset = dataset['train'].map(load_image_train, num_parallel_calls=tf.data.AUTOTUNE)
train_batches = train_dataset.cache().shuffle(BUFFER_SIZE).batch(BATCH_SIZE).repeat()
train_batches = train_batches.prefetch(buffer_size=tf.data.AUTOTUNE)

# Test pipeline (unbatched, for visualization)
test_dataset = dataset['test'].map(load_image_test)
# Test pipeline (batched, for validation during training)
test_batches = test_dataset.batch(BATCH_SIZE)

print("\nData pipelines created successfully.")

# =============================
# SECTION 3: Build the U-Net Model
# =============================
# This is a standard U-Net architecture.

from tensorflow.keras import layers

def double_conv_block(x, n_filters):
    """A block of two 3x3 convolution layers"""
    x = layers.Conv2D(n_filters, 3, padding="same", activation="relu", kernel_initializer="he_normal")(x)
    x = layers.Conv2D(n_filters, 3, padding="same", activation="relu", kernel_initializer="he_normal")(x)
    return x

def downsample_block(x, n_filters):
    """Downsamples with max-pooling and applies a double-conv block"""
    f = double_conv_block(x, n_filters)
    p = layers.MaxPool2D(2)(f)
    p = layers.Dropout(0.3)(p)
    return f, p

def upsample_block(x, conv_features, n_filters):
    """Upsamples with transposed conv and concatenates skip connection"""
    x = layers.Conv2DTranspose(n_filters, 3, 2, padding="same")(x)
    x = layers.concatenate([x, conv_features])
    x = layers.Dropout(0.3)(x)
    x = double_conv_block(x, n_filters)
    return x

def build_unet_model():
    """Builds the U-Net model."""
    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))

    # Encoder (Downsampling Path)
    f1, p1 = downsample_block(inputs, 64)
    f2, p2 = downsample_block(p1, 128)
    f3, p3 = downsample_block(p2, 256)
    f4, p4 = downsample_block(p3, 512)

    # Bottleneck
    bottleneck = double_conv_block(p4, 1024)

    # Decoder (Upsampling Path) with Skip Connections
    u6 = upsample_block(bottleneck, f4, 512)
    u7 = upsample_block(u6, f3, 256)
    u8 = upsample_block(u7, f2, 128)
    u9 = upsample_block(u8, f1, 64)

    # Output layer
    # We have 3 classes (pet, border, background)
    outputs = layers.Conv2D(3, 1, padding="same", activation="softmax")(u9)

    unet_model = tf.keras.Model(inputs, outputs, name="U-Net")
    return unet_model

model = build_unet_model()
model.summary()

# =============================
# SECTION 4: Train the Model
# =============================
print("\nStarting model training...")

# We use SparseCategoricalCrossentropy because our masks are not one-hot encoded
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              metrics=['accuracy'])

# Calculate steps per epoch
TRAIN_LENGTH = info.splits['train'].num_examples
STEPS_PER_EPOCH = TRAIN_LENGTH // BATCH_SIZE
VALIDATION_STEPS = info.splits['test'].num_examples // BATCH_SIZE // 10 # Validate on 10% of test set

# Train for 20 epochs (can be increased, but this is good for a demo)
EPOCHS = 20

history = model.fit(
    train_batches,
    epochs=EPOCHS,
    steps_per_epoch=STEPS_PER_EPOCH,
    validation_steps=VALIDATION_STEPS,
    validation_data=test_batches
)

print("\nTraining complete.")

# =============================
# SECTION 5: Evaluation & Visuals
# =============================

print("\nDisplaying training history (Loss & Accuracy):")
# Plot training & validation accuracy and loss
loss = history.history['loss']
val_loss = history.history['val_loss']
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.legend()
plt.title('Loss')
sns.despine()

plt.subplot(1, 2, 2)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend()
plt.title('Accuracy')
sns.despine()

plt.tight_layout()
plt.show()


# --- Show Predictions (The "Show-Off" Step) ---
print("\nDisplaying sample predictions...")

def create_mask(pred_mask):
    """Converts the model's softmax output into a solid mask"""
    pred_mask = tf.argmax(pred_mask, axis=-1)
    pred_mask = pred_mask[..., tf.newaxis]
    return pred_mask[0]

def display_predictions(dataset, num=3):
    """Displays the original image, true mask, and predicted mask"""
    plt.figure(figsize=(15, num * 5))

    # `dataset` should be UNBATCHED here
    for i, (image, mask) in enumerate(dataset.take(num)):
        # Make a prediction, adding the batch dimension with tf.newaxis
        pred_mask = model.predict(image[tf.newaxis, ...])

        # Plot Original Image
        plt.subplot(num, 3, i*3 + 1)
        plt.title("Input Image")
        plt.imshow(image)
        plt.axis('off')

        # Plot True Mask
        plt.subplot(num, 3, i*3 + 2)
        plt.title("True Mask")
        plt.imshow(tf.keras.utils.array_to_img(mask))
        plt.axis('off')

        # Plot Predicted Mask
        plt.subplot(num, 3, i*3 + 3)
        plt.title("Predicted Mask")
        plt.imshow(tf.keras.utils.array_to_img(create_mask(pred_mask)))
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# Create a project directory to save model
PROJECT_DIR = "/content/unet_project"
os.makedirs(PROJECT_DIR, exist_ok=True)

# --- THIS LINE HAS BEEN FIXED ---
# Take 3 samples from the *unbatched* test dataset and show the results
display_predictions(test_dataset.shuffle(100))

# =============================
# SECTION 6: Download Artifacts
# =============================
print("\n============================================")
print("SAVING & DOWNLOADING YOUR TRAINED MODEL")

# Save the model in Keras's native format
MODEL_SAVE_PATH = os.path.join(PROJECT_DIR, "unet_pet_model.keras")
model.save(MODEL_SAVE_PATH)
print(f"Model saved to: {MODEL_SAVE_PATH}")

# Create a download link for the model
# display(FileLink(MODEL_SAVE_PATH)) # Comment out or remove if not using google.colab.files
print("============================================")

# =============================
# SECTION 7: Usage & README (as variable)
# =============================
usage = """
How to Use:
-----------
1. Run the entire notebook (Runtime -> Run all). No API key is needed.
2. The notebook will:
    - Install tensorflow and tensorflow-datasets.
    - Download the 'oxford_iiit_pet' dataset automatically.
    - Build a U-Net model from scratch.
    - Train the U-Net model for 20 epochs.
    - Display training graphs (Loss, Accuracy).
    - Show visual predictions (Image, True Mask, Predicted Mask) on test images.
    - Provide a download link for your trained model (`unet_pet_model.keras`).
"""

# To view the usage instructions at any time:
# print(usage)
```
